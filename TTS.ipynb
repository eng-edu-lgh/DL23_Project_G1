```python
#@markdown ðŸ”„ [1] Making functions (sample): tts(text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def tts(text):
  gtts_object = gTTS(text,lang = "en", slow = True) 
  gtts_object.save("myaudio.mp3")
  return Audio("myaudio.mp3")
     
```
